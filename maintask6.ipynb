```python
df=df.drop_duplicates()
df.shape
```

```python
df.drop(['PatientID','DoctorInCharge'],axis=1 ,inplace=True)
df.columns
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR  =Q3-Q1
print(IQR)
```

```python
df=df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]
df
```

```python
questions = ["1.How many people have diabetes and how many people don't have diabetes ?",
             "2.How many Male and Female have Diabetes ?",
             "3.How many smoking people have diabetes ? ",
             "4.How many Educational Level people have Diabetes ?",
             "5.How many people get quality water and sill have Diabetes for Gender ?"]
questions
```

```python
# Question one answer
# How  many people have diabetes and how many people don't have diabetes ?

df.Diagnosis.value_counts()
```

```python
# Bar Chart

df.Diagnosis.value_counts().plot(kind='bar',color=['orchid','salmon'])
plt.title('Diabetes value counts')
plt.xlabel('1=Dibetes,0=Not Diabetes')
plt.ylabel('Value Counts');
```

```python
# Question Two
# How many Male and Female have Diabetes ?
# 0 - Male
# 1 - Female

df.Gender.value_counts()
```

```python
df.Gender.value_counts().plot(kind='pie',figsize=(8,6))
plt.title('Male Female Ratio')
plt.legend(['Male','Female']);
```

```python
pd.crosstab(df.Diagnosis , df.Gender)
```

```python
pd.crosstab(df.Diagnosis,df.Gender).plot(kind='bar',color=['lightskyblue','plum'])
plt.title('Diabetes According to Gender')
plt.xlabel("0=Male,1=Female")  ;                                                         
```

```python
# Question three
# How many Smoking people have Diabetes ?

pd.crosstab(df.Diagnosis , df.Smoking)
```

```python
pd.crosstab(df.Diagnosis , df.Smoking).plot(kind='bar',color=['coral','khaki'])
plt.title('How many Smoking people has diabetes')
plt.xlabel('0=Non Smoking ,1=Smoking');                                            
```

```python
# Question four
# How many Educational Level people have Diabetes ?

pd.crosstab(df.Diagnosis , df.EducationLevel)
```

```python
pd.crosstab(df.Diagnosis , df.EducationLevel).plot(kind='bar',color=['coral','lightskyblue','plum','khaki'])
plt.title('Education For Diabetes People')
plt.xlabel('Education Level');
```

```python
# Question Five
# How many Daibetes people get drinkable or not drinkable water ?
# 0 = Drinkable , 1=NotDrinkable

pd.crosstab(df.Gender,df.WaterQuality)
```

```python
pd.crosstab(df.Gender,df.WaterQuality).plot(kind='bar',color=['lightskyblue','plum'])
plt.title('Water Quality for Diabetes Peolple')
plt.xlabel("0=Drinkable,1=Not Drinkable")  ;  
```
